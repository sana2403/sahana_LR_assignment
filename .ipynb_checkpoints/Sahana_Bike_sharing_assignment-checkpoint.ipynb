# Bike Sharing Assignment

#### Problem Statement

A consulting company to understand the factors on which the demand for these shared bikes depends. Specifically, they want to understand the factors affecting the demand for these shared bikes in the American market. The company wants to know:

- Which variables are significant in predicting the demand for shared bikes.
- How well those variables describe the bike demands

# Data Modelling

```python
bike.head()
```

```python
bike.shape
```

```python
bike.info()
```

#### Splitting the data into Training and Testing Sets

```python
df_train, df_test = train_test_split(bike, train_size=0.7, random_state=100)
```

```python
print(df_train.shape)
print(df_test.shape)
```

### Rescaling the Features
 - Using MinMax Scaling

```python
# Create an instance of minmax scaler
scaler = MinMaxScaler()
```

```python
# Apply MinMaxScaler on numerical variales
num_vars = ['temp', 'hum', 'windspeed', 'cnt']

df_train[num_vars] = scaler.fit_transform(df_train[num_vars])

df_train.head()
```

#### Create X and y sets for the model building

```python
y_train = df_train.pop('cnt')
X_train = df_train
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
y_train.head()
```

## Building our model using RFE 

- Recursive feature elimination

```python
# Running RFE with the output number of the variable equal to 20

lm = LinearRegression()
lm.fit(X_train,y_train)

# Running RFE
rfe = RFE(lm, n_features_to_select=15)
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns, rfe.support_, rfe.ranking_))
```

```python
# RFE selected features
col = X_train.columns[rfe.support_]
col
```

```python
# Features Rejected by the RFE
X_train.columns[~(rfe.support_)]
```

```python
# Creating X train with RFE Selected Variables
X_train_rfe = X_train[col]
```

```python
# Adding Constant
X_train_rfe = sm.add_constant(X_train_rfe)

# Create Object for the Model
lm = sm.OLS(y_train, X_train_rfe).fit()

# View Summary of Statistics
lm.summary()
```

- As the statistic says there are strong multicollinearity, we have to check VIF(variance inflation factor)

```python
vif = pd.DataFrame()
vif['Feature'] = X_train_rfe.columns
vif['VIF'] = [variance_inflation_factor(X_train_rfe.values, i) for i in range(X_train_rfe.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by='VIF', ascending=False)
vif
```

```python
# As VIF is infinity for holiday we can drop it
X_train_rfe = X_train_rfe.drop('holiday', axis=1)
X_train_rfe.columns
```

```python
# Add constant to all the Variables
X_train_sm = sm.add_constant(X_train_rfe)
# Create the Model
lr = sm.OLS(y_train, X_train_sm)
# Fit the Model
lr_model = lr.fit()
# View the Parameter of the Model
lr_model.summary()
```

```python
vif = pd.DataFrame()
vif['Feature'] = X_train_rfe.columns
vif['VIF'] = [variance_inflation_factor(X_train_rfe.values, i) for i in range(X_train_rfe.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by='VIF', ascending=False)
vif
```

```python
# As VIF is high for const we can drop it
X_train_rfe = X_train_rfe.drop('workingday', axis=1)
X_train_rfe.columns
```

```python
# Add constant to all the Variables
X_train_sm = sm.add_constant(X_train_rfe)
# Create the Model
lr = sm.OLS(y_train, X_train_sm)
# Fit the Model
lr_model = lr.fit()
# View the Parameter of the Model
lr_model.summary()
```

```python
vif = pd.DataFrame()
vif['Feature'] = X_train_rfe.columns
vif['VIF'] = [variance_inflation_factor(X_train_rfe.values, i) for i in range(X_train_rfe.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by='VIF', ascending=False)
vif
```

```python
X_train_rfe.shape
```

```python
X_train_rfe = X_train_rfe.drop('Saturday', axis=1)
X_train_rfe.columns
```

- Dropping off Saturday as its P-value if higher than 0.05

```python
# Add constant to all the Variables
X_train_sm = sm.add_constant(X_train_rfe)
# Create the Model
lr = sm.OLS(y_train, X_train_sm)
# Fit the Model
lr_model = lr.fit()
# View the Parameter of the Model
lr_model.summary()
```

```python
vif = pd.DataFrame()
vif['Feature'] = X_train_rfe.columns
vif['VIF'] = [variance_inflation_factor(X_train_rfe.values, i) for i in range(X_train_rfe.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by='VIF', ascending=False)
vif
```

## Observation on above model

- **We have P(F-statistic) is very low (equal to Zero) which tells overall model is good fit**
- **Adjusted R2 score is 83.9% which is good**
- **P-value for all the variables are less then 0.05 ,which statistically significant**
- **VIF value for all variable is less than 5**

## Making Predictions

### Scaling on test set

```python
# Apply MinMaxScaler on numerical variales
num_vars = ['temp', 'hum', 'windspeed', 'cnt']

df_test[num_vars] = scaler.transform(df_test[num_vars])

df_test.head()
```

```python
# Create X_test and y_test
y_test = df_test.pop('cnt')
X_test = df_test
```

```python
print(y_test.shape)
print(X_test.shape)
```

```python
y_test.head()
```

```python
# Add constant
X_test_sm = sm.add_constant(X_test)
X_test_sm.head()
```

```python
X_test_sm.columns
```

```python
# Drop features which are removed during training process
X_test_sm = X_test_sm.drop(['atemp', 'holiday', 'workingday', 'spring', 'Monday', 'Thursday', 'Tuesday', 'Wednesday', 'Saturday','Dec',
       'Feb', 'Jan', 'Jul', 'Jun', 'Mar', 'May', 'Nov'], axis=1)
```

```python
X_test_sm.info()
```

```python
y_test_pred = lr_model.predict(X_test_sm)
```

```python
y_test_pred.head()
```

```python
# Plotting y_test and y_pred
plt.figure(figsize=(10,5))
sns.regplot(x=y_test, y=y_test_pred)
plt.title('Actual vs. Predicted Rent of Bikes Count')
plt.xlabel('Actual Rent of Bikes Count')
plt.ylabel('Predicted Rent of Bikes Count')
plt.show()
```

**Above Scatter Plot Shows the Linear Relationship between Actual Rent of Bikes Count and Predicted Rent of Bikes Count.**

```python
r2_score(y_true=y_test, y_pred=y_test_pred)
```

**The final model accuracy on test set is 81.14% which is pretty good.**

**The Model accuraacy between Train set and Test set are less than 5% so it is acceptable and shows good fit regression line between actual and predicted rent of Bikes count**

```python
# The coefficients values which affects more on rent bike counts
Cofficients = round(lr_model.params,2)
coef = Cofficients.sort_values(ascending = False)
coef
```

**Equation of the Best fit line**

**cnt = 0.53 * temp + 0.23 * yr + 0.13 * winter + 0.12 * Sep + 0.11 * summer + 0.06 * Aug + 0.04 * Oct -0.04 * Sunday - 0.05 *           Mist - 0.17 * hum -0.19 * windspeed -0.24 * light_snow**

# Interpretation of the model

> **Temperature is the Most Significant Feature which affects the Business positively.**
  

> **Whereas the other Environmental condition such as Raining, Humidity, Windspeed and Cloudy affects the Business negatively.**

  
> **The Demand of Bikes is more in the Winter and Summer season, mostly user don't like to travel using Bikes in Rainy Day or       Rainy Season and when there is snowing.**
